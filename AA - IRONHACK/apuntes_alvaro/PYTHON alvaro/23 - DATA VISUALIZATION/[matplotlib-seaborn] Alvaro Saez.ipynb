```python
#Matplotlib

# MEDIAN STEPS BY MONTH_ENCODED
fitbit_pt_sbm = fitbit_db.pivot_table(index =["Months_encoded"],values =["Steps"])
fitbit_pt_sbm["Steps"] = fitbit_pt_sbm["Steps"].round(0).astype(int)
# your code here
fig_ch2_sbm, ax_ch2_sbm =  plt.subplots(nrows=1, ncols=1,figsize=(12, 4))
ax_ch2_sbm.bar(fitbit_pt_sbm.index.tolist(),fitbit_pt_sbm["Steps"].tolist(), label='steps')
ax_ch2_sbm.legend()
ax_ch2_sbm.set_xlabel("MONTHS")
ax_ch2_sbm.set_ylabel("MEDIAN STEPS")
ax_ch2_sbm.set_title("MEDIAN STEPS BY MONTH_ENCODED")

plt.savefig("../figures/median_steps_by_month.png")
plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
fitbit_pt_sbw = fitbit_db.pivot_table(index =["Work_or_Weekend"],values =["Steps"])
fitbit_pt_sbw["Steps"] = fitbit_pt_sbw["Steps"].round(0).astype(int)
fitbit_pt_sbw.index = np.where(fitbit_pt_sbw.index == 0, "weekend", "week")

fig_ch2_sbw, ax_ch2_sbw =  plt.subplots(nrows=1, ncols=1,figsize=(12, 4))
ax_ch2_sbw.barh(fitbit_pt_sbw.index.tolist(),fitbit_pt_sbw["Steps"].tolist(), label='steps')
ax_ch2_sbw.legend()
ax_ch2_sbw.set_xlabel("MEDIAN STEPS")
ax_ch2_sbw.set_title("MEDIAN STEPS BY WORK_OR_WEEKEND")
ax_ch2_sbw.text(fitbit_pt_sbw["Steps"][0],"weekend",str(fitbit_pt_sbw["Steps"][0]), color="black")
ax_ch2_sbw.text(fitbit_pt_sbw["Steps"][1],"week",str(fitbit_pt_sbw["Steps"][1]), color="black")

plt.savefig("../figures/week_or_week_end.png")
plt.show()
```
