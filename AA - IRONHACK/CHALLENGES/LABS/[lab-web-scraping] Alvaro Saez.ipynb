#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
# This is the url you will scrape in this exercise
url_wiki_py ='https://en.wikipedia.org/wiki/Python' 
````

````python
# your code here
response_wiki_py = requests.get(url_wiki_py).content
parsed_wiki_py = bs4.BeautifulSoup(response_wiki_py,"html.parser").find("div", {"id": "bodyContent"}).find_all("a")
parsed_wiki_py_list = [str(i.string)+ ": " + str(i["href"]) for i in parsed_wiki_py]
#parsed_wiki_py_list
def wiki_py(x):
    splited_0 = x.split(": ")[0]
    splited_1 = x.split(": ")[1]
    if "https:" in splited_1:
        return splited_0 + ": " + splited_1
    else:
        return splited_0 + ": https://en.wikipedia.org/wiki/Python" + splited_1
parsed_final_wiki_py_list = list(map(wiki_py,parsed_wiki_py_list))
parsed_final_wiki_py_list
````

#### Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# This is the url you will scrape in this exercise 
url_imdb250 = 'https://www.imdb.com/chart/top'
response_imdb250 = requests.get(url_imdb250).content
````

````python
# your code here
parsed_imdb250_name = bs4.BeautifulSoup(response_imdb250,"html.parser").find_all("table", {"class": "chart full-width"})[0].find_all("a")
parsed_imdb250_name_list = [i.string for i in parsed_imdb250_name if i.string !=None]

parsed_imdb250_href = bs4.BeautifulSoup(response_imdb250,"html.parser").find_all("table", {"class": "chart full-width"})[0].find_all("a")
parsed_imdb250_href_list = ["https://www.imdb.com/"+i["href"] for i in parsed_imdb250_href if i.string !=None]

parsed_imdb250_director = bs4.BeautifulSoup(response_imdb250,"html.parser").find_all("table", {"class": "chart full-width"})[0].find_all("a")
parsed_imdb250_director_list = [i["title"].split("(dir.)")[0].strip() for i in parsed_imdb250_director if i.string !=None]

parsed_imdb250_year = bs4.BeautifulSoup(response_imdb250,"html.parser").find_all("table", {"class": "chart full-width"})[0].find_all("span")
parsed_imdb250_year_list = [i.string for i in parsed_imdb250_year if i.string !=None and i.string !="\xa0" ]

parsed_imdb250_stars = bs4.BeautifulSoup(response_imdb250,"html.parser").find_all("td", {"class": "ratingColumn imdbRating"})
parsed_imdb250_stars_list = [i.strong.string for i in parsed_imdb250_stars]

parsed_imdb250_df = pd.DataFrame({"movie name":parsed_imdb250_name_list, "initial release":parsed_imdb250_year_list,"director name":parsed_imdb250_director_list,"stars":parsed_imdb250_stars_list,"url":parsed_imdb250_href_list})
parsed_imdb250_df
````
