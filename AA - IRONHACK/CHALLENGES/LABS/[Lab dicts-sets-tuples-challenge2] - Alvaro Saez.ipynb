#### What is the relationship among the following values:

* len(set1)
* len(set2)
* len(set3)
* len(set4)
* len(set5)

Use a math formular to represent that relationship. Test your formular with Python code.

````python
# Your code here
print(len(set1))
print(len(set2))
print(len(set3))
print(len(set4))
print(len(set5))

# Your code here
    #set3 comes from difference between 1 and 2
    #set4 comes from difference between 2 and 1
    #set5 comes from intersection between 1 and 2
    
#we will store the lenght in new variables to calculate their relationship with some statistics methods
len_set1 = len(set1)
len_set2 = len(set2)
len_set3 = len(set3)
len_set4 = len(set4)
len_set5 = len(set5)

import numpy as np

mean_sets = np.mean([len_set1, len_set2, len_set3, len_set4, len_set5])
print("mean of sets is: ",mean_sets)

var_sets = np.var([len_set1, len_set2, len_set3, len_set4, len_set5])
print("variance of sets is: ",var_sets)

std_sets = np.std([len_set1, len_set2, len_set3, len_set4, len_set5])
print("standard deviation of sets is: ",std_sets)
````
